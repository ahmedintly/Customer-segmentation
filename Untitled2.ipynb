```python
from flask import Flask, request, render_template, redirect, url_for, session
import torch
import csv
import os
import glob
import shutil
from datetime import date
import sqlite3
from werkzeug.security import generate_password_hash, check_password_hash
from pyngrok import ngrok


# ================== APP ==================
app = Flask(__name__)
app.secret_key = "change-this-secret"

MODEL_PATH = "/content/last (2).pt"
CSV_PATH   = "/content/caloris.csv"   # Ù„Ùˆ Ù…Ù„ÙÙƒ ÙÙŠ Ù…ÙƒØ§Ù† ØªØ§Ù†ÙŠ ØºÙŠÙ‘Ø± Ø§Ù„Ù…Ø³Ø§Ø± Ù‡Ù†Ø§
DB_PATH    = "/content/app.db"

# ================== NORMALIZE LABELS ==================
def norm_label(s: str) -> str:
    s = (s or "").strip().lower()
    s = s.replace("_", " ")
    s = " ".join(s.split())
    return s

# ================== DATABASE ==================
def db_conn():
    return sqlite3.connect(DB_PATH)

def init_db():
    con = db_conn()
    cur = con.cursor()

    cur.execute(
        "CREATE TABLE IF NOT EXISTS users ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "name TEXT NOT NULL,"
        "email TEXT NOT NULL UNIQUE,"
        "password_hash TEXT NOT NULL"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS profiles ("
        "user_id INTEGER PRIMARY KEY,"
        "sex TEXT,"
        "age INTEGER,"
        "height_cm REAL,"
        "weight_kg REAL,"
        "activity TEXT,"
        "goal TEXT,"
        "daily_goal REAL,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    con.commit()
    con.close()

init_db()

def get_user_by_email(email: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT id, name, email, password_hash FROM users WHERE email=?", (email,))
    row = cur.fetchone()
    con.close()
    return row

def create_user(name: str, email: str, password: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO users (name,email,password_hash) VALUES (?,?,?)",
        (name, email, generate_password_hash(password))
    )
    con.commit()
    uid = cur.lastrowid
    con.close()
    return uid

def upsert_profile(uid: int, sex: str, age: int, height: float, weight: float, activity: str, goal: str, daily: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO profiles VALUES (?,?,?,?,?,?,?,?) "
        "ON CONFLICT(user_id) DO UPDATE SET "
        "sex=excluded.sex, age=excluded.age, height_cm=excluded.height_cm, "
        "weight_kg=excluded.weight_kg, activity=excluded.activity, "
        "goal=excluded.goal, daily_goal=excluded.daily_goal",
        (uid, sex, age, height, weight, activity, goal, daily)
    )
    con.commit()
    con.close()

def get_daily_goal(uid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT daily_goal FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return row[0] if row else None

# ================== MODEL ==================
model = torch.hub.load(
    "ultralytics/yolov5",
    "custom",
    path=MODEL_PATH,
    trust_repo=True
)

# ================== LOAD CALORIES CSV ==================
# Ù†Ø®Ø²Ù†Ù‡Ø§ Ø¨Ù†Ø³Ø®Ø© normalized Ø¹Ù„Ø´Ø§Ù† ØªØ·Ø§Ø¨Ù‚ YOLO Ù…Ù‡Ù…Ø§ Ø§Ø®ØªÙ„ÙØª Ø§Ù„ØµÙŠØºØ©
calorie_data = {}   # key: normalized -> kcal per 100g
label_display = {}  # key: normalized -> original label from CSV (Ù„Ù„ÙˆØ§Ø¬Ù‡Ø©)

with open(CSV_PATH, "r", encoding="cp1256", newline="") as f:
    reader = csv.reader(f)
    for row in reader:
        if not row or len(row) < 3:
            continue
        raw_name = row[0].strip()
        key = norm_label(raw_name)
        try:
            calorie_data[key] = float(row[2])
            label_display[key] = raw_name
        except:
            pass

# ================== CALC DAILY GOAL (BMR/TDEE) ==================
def calc_daily(sex, age, height, weight, activity, goal):
    sex = (sex or "").lower().strip()

    if sex == "male":
        bmr = 10*weight + 6.25*height - 5*age + 5
    else:
        bmr = 10*weight + 6.25*height - 5*age - 161

    factors = {
        "sedentary": 1.2,
        "light": 1.375,
        "moderate": 1.55,
        "active": 1.725,
        "very_active": 1.9
    }
    tdee = bmr * factors.get(activity, 1.2)

    if goal == "lose":
        tdee -= 500
    elif goal == "gain":
        tdee += 300

    # Ø­Ø¯ Ø£Ù…Ø§Ù† Ø¨Ø³ÙŠØ·
    tdee = max(tdee, 1200 if sex == "female" else 1500)
    return float(round(tdee, 0))

# ================== DAILY SESSION STATE ==================
def ensure_today():
    today = str(date.today())
    if session.get("day") != today:
        session["day"] = today
        session["meals"] = []     # [{meal_type,total,details}]
        session["consumed"] = 0.0

def require_login():
    return "user_id" in session

# ================== ROUTES ==================
@app.route("/")
def root():
    if not require_login():
        return redirect(url_for("login"))

    daily = get_daily_goal(session["user_id"])
    if daily is None:
        return redirect(url_for("onboarding"))

    session["daily_goal"] = float(daily)
    ensure_today()
    return redirect(url_for("dashboard"))

# ---------- AUTH ----------
@app.route("/signup", methods=["GET", "POST"])
def signup():
    if request.method == "POST":
        name = request.form.get("name", "").strip()
        email = request.form.get("email", "").strip().lower()
        password = request.form.get("password", "")

        if not name or not email or not password:
            return render_template("signup.html", error="Ø§Ù…Ù„Ø£ ÙƒÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª.")

        if get_user_by_email(email):
            return render_template("signup.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ Ù…ÙˆØ¬ÙˆØ¯ Ø¨Ø§Ù„ÙØ¹Ù„. Ø³Ø¬Ù„ Ø¯Ø®ÙˆÙ„.")

        uid = create_user(name, email, password)
        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = email

        return redirect(url_for("onboarding"))

    return render_template("signup.html")

@app.route("/login", methods=["GET", "POST"])
def login():
    if request.method == "POST":
        email = request.form.get("email", "").strip().lower()
        password = request.form.get("password", "")

        row = get_user_by_email(email)
        if not row:
            return render_template("login.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ ØºÙŠØ± ØµØ­ÙŠØ­.")

        uid, name, em, pw_hash = row
        if not check_password_hash(pw_hash, password):
            return render_template("login.html", error="ÙƒÙ„Ù…Ø© Ø§Ù„Ù…Ø±ÙˆØ± ØºÙŠØ± ØµØ­ÙŠØ­Ø©.")

        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = em

        return redirect(url_for("root"))

    return render_template("login.html")

@app.route("/logout")
def logout():
    session.clear()
    return redirect(url_for("login"))

# ---------- ONBOARDING ----------
@app.route("/onboarding", methods=["GET", "POST"])
def onboarding():
    if not require_login():
        return redirect(url_for("login"))

    if request.method == "POST":
        sex = request.form.get("sex")
        age = int(request.form.get("age"))
        height = float(request.form.get("height_cm"))
        weight = float(request.form.get("weight_kg"))
        activity = request.form.get("activity")
        goal = request.form.get("goal")

        daily = calc_daily(sex, age, height, weight, activity, goal)
        upsert_profile(session["user_id"], sex, age, height, weight, activity, goal, daily)

        session["daily_goal"] = daily
        ensure_today()
        return redirect(url_for("dashboard"))

    return render_template("onboarding.html")

# ---------- DASHBOARD ----------
@app.route("/dashboard")
def dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    daily = float(session["daily_goal"])
    consumed = float(session.get("consumed", 0.0))

    remaining = max(0.0, round(daily - consumed, 2))  # âœ… Ù„Ø§ Ø³Ø§Ù„Ø¨
    over_by = max(0.0, round(consumed - daily, 2))    # âœ… ÙØ±Ù‚ Ø§Ù„ØªØ¬Ø§ÙˆØ²

    return render_template(
        "dashboard.html",
        user_name=session.get("user_name"),
        daily_goal=round(daily, 0),
        consumed=round(consumed, 2),
        remaining=remaining,
        over_by=over_by,
        meals=session.get("meals", []),
        day=session.get("day")
    )

# ---------- MEALS ----------
@app.route("/start/<meal>")
def start_meal(meal):
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    meal = (meal or "").lower().strip()
    allowed = {"breakfast", "lunch", "dinner", "snack"}
    if meal not in allowed:
        meal = "breakfast"

    session["current_meal"] = meal
    return redirect(url_for("upload"))

@app.route("/upload", methods=["GET", "POST"])
def upload():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    meal = session.get("current_meal", "breakfast")

    if request.method == "POST":
        if "image" not in request.files or request.files["image"].filename == "":
            return render_template("upload.html", error="Ø§Ø±ÙØ¹ ØµÙˆØ±Ø© Ù…Ù† ÙØ¶Ù„Ùƒ.", meal_type=meal)

        img = request.files["image"]
        image_path = "/content/static/uploaded.jpg"
        img.save(image_path)

        results = model(image_path)
        df = results.pandas().xyxy[0]

        foods_raw = df["name"].unique().tolist() if len(df) else []
        foods = [norm_label(x) for x in foods_raw]  # âœ… normalize YOLO output

        # save annotated
        annotated_url = None
        try:
            results.save()  # runs/detect/exp*
            exp_dirs = sorted(glob.glob("runs/detect/exp*"), key=os.path.getmtime)
            latest = exp_dirs[-1] if exp_dirs else None
            if latest:
                imgs = glob.glob(os.path.join(latest, "*.jpg")) + glob.glob(os.path.join(latest, "*.png"))
                if imgs:
                    shutil.copy(imgs[0], "/content/static/annotated.jpg")
                    annotated_url = "/static/annotated.jpg"
        except:
            annotated_url = None

        # build display results (per 100g)
        res = []
        for k in foods:
            if k in calorie_data:
                res.append(f"{label_display.get(k, k)}: {calorie_data[k]} kcal / 100g")
            else:
                res.append(f"{label_display.get(k, k)}: Ù…Ø´ Ù…ÙˆØ¬ÙˆØ¯ ÙÙŠ Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª")

        return render_template(
            "results.html",
            foods=foods,                 # âœ… normalized keys
            res=res,
            annotated_url=annotated_url,
            daily_goal=session["daily_goal"],
            meal_type=meal,
            label_display=label_display  # âœ… for UI names
        )

    return render_template("upload.html", meal_type=meal)

@app.route("/calculate", methods=["POST"])
def calculate():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    foods = request.form.getlist("food")       # âœ… normalized keys from form
    qtys  = request.form.getlist("quantity")

    total = 0.0
    details = []

    for f, q in zip(foods, qtys):
        key = norm_label(f)
        q = float(q)

        cal_per_100 = calorie_data.get(key, 0.0)
        cal = cal_per_100 * (q / 100.0)

        total += cal
        details.append({
            "food": label_display.get(key, key),
            "qty": q,
            "cal": round(cal, 2) if cal_per_100 > 0 else None
        })

    total = round(total, 2)

    session["meals"] = session.get("meals", [])
    session["meals"].append({
        "meal_type": session.get("current_meal", "meal"),
        "total": total,
        "details": details
    })

    session["consumed"] = float(session.get("consumed", 0.0)) + total

    return redirect(url_for("dashboard"))
ngrok.kill()
public_url = ngrok.connect(5000)
print("âœ… Ø§ÙØªØ­ Ø§Ù„Ù„ÙŠÙ†Ùƒ Ø¯Ù‡:", public_url)
# ================== RUN ==================
if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)

```

```python
import cv2
import numpy as np
import gradio as gr
import mediapipe as mp

mp_pose = mp.solutions.pose
mp_draw = mp.solutions.drawing_utils
mp_selfie = mp.solutions.selfie_segmentation

POSES = ["WarriorII", "Tree", "Plank", "DownDog", "Squat"]  # Ù…Ø«Ø§Ù„ ÙƒØ¨Ø¯Ø§ÙŠØ©

def angle(a, b, c):
    a = np.array(a); b = np.array(b); c = np.array(c)
    ba = a - b
    bc = c - b
    cosang = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc) + 1e-6)
    cosang = np.clip(cosang, -1.0, 1.0)
    return float(np.degrees(np.arccos(cosang)))

def in_range(val, target, tol):
    return abs(val - target) <= tol

# ====== Rules Ø¨Ø³ÙŠØ·Ø© ÙƒØ¨Ø¯Ø§ÙŠØ© (ØªÙ‚Ø¯Ø± ØªØ²ÙˆØ¯ ÙˆØªØ¸Ø¨Ø·) ======
# ÙƒÙ„ Pose Ù„Ù‡Ø§ Ø´Ø±ÙˆØ· Ø²ÙˆØ§ÙŠØ§ ØªÙ‚Ø±ÙŠØ¨ÙŠØ©
def pose_rules(pose_name, pts):
    # pts dict: contains keypoints in pixels
    # return (ok:bool, feedback:str)
    # key names: l_sho,l_elb,l_wri,r_sho,r_elb,r_wri,l_hip,l_knee,l_ank,r_hip,r_knee,r_ank
    if pose_name == "Plank":
        # ÙƒØªÙ-ÙˆØ±Ùƒ-ÙƒØ§Ø­Ù„ Ù‚Ø±ÙŠØ¨ Ù…Ù† Ø®Ø· Ù…Ø³ØªÙ‚ÙŠÙ… (Ø²Ø§ÙˆÙŠØ© Ø§Ù„ÙˆØ±Ùƒ ~ 170)
        hip = angle(pts["r_sho"], pts["r_hip"], pts["r_ank"])
        ok = in_range(hip, 175, 15)
        return ok, f"Hip angle={hip:.0f} (target~175)"
    if pose_name == "Squat":
        knee = angle(pts["r_hip"], pts["r_knee"], pts["r_ank"])
        ok = in_range(knee, 90, 20)
        return ok, f"Knee angle={knee:.0f} (target~90)"
    if pose_name == "Tree":
        # Ø±ÙƒØ¨Ø© Ø±Ø¬Ù„ Ù…Ø«Ù†ÙŠØ© (Ø²Ø§ÙˆÙŠØ© Ø±ÙƒØ¨Ø© ØµØºÙŠØ±Ø©/Ù…ØªÙˆØ³Ø·Ø©) + ØªÙˆØ§Ø²Ù† ØµØ¹Ø¨ Ø¨Ø§Ù„Ù‚ÙˆØ§Ø¹Ø¯ Ø¨Ø³
        knee = angle(pts["l_hip"], pts["l_knee"], pts["l_ank"])
        ok = knee < 140  # ØªÙ‚Ø±ÙŠØ¨
        return ok, f"Left knee angle={knee:.0f} (want bent)"
    if pose_name == "DownDog":
        # Ø´ÙƒÙ„ V: Ø²Ø§ÙˆÙŠØ© Ø§Ù„ÙˆØ±Ùƒ Ø£ØµØºØ± Ù…Ù† Ø§Ù„Ø¨Ù„Ø§Ù†Ùƒ ØºØ§Ù„Ø¨Ù‹Ø§ ~ 90-120
        hip = angle(pts["r_sho"], pts["r_hip"], pts["r_ank"])
        ok = in_range(hip, 110, 25)
        return ok, f"Hip angle={hip:.0f} (target~110)"
    if pose_name == "WarriorII":
        # Ø±ÙƒØ¨Ø© Ø£Ù…Ø§Ù…ÙŠØ© ~90 ÙˆØ°Ø±Ø§Ø¹ÙŠÙ† Ù…ÙØ±ÙˆØ¯ÙŠÙ† ~170
        knee = angle(pts["r_hip"], pts["r_knee"], pts["r_ank"])
        l_elb = angle(pts["l_sho"], pts["l_elb"], pts["l_wri"])
        r_elb = angle(pts["r_sho"], pts["r_elb"], pts["r_wri"])
        ok = in_range(knee, 90, 20) and in_range(l_elb, 170, 25) and in_range(r_elb, 170, 25)
        return ok, f"Knee={knee:.0f}, LElb={l_elb:.0f}, RElb={r_elb:.0f}"
    return False, "No rules for this pose yet."

pose = mp_pose.Pose(model_complexity=1, enable_segmentation=False,
                    min_detection_confidence=0.5, min_tracking_confidence=0.5)
seg = mp_selfie.SelfieSegmentation(model_selection=1)

def process(frame, pose_name):
    if frame is None:
        return None

    # frame from gradio is RGB numpy
    img = frame.copy()
    h, w = img.shape[:2]

    rgb = img
    res = pose.process(rgb)

    # segmentation mask for contour-ish highlight (Ø§Ø®ØªÙŠØ§Ø±ÙŠ)
    seg_res = seg.process(rgb)
    mask = seg_res.segmentation_mask
    mask_bin = (mask > 0.3).astype(np.uint8) * 255
    edges = cv2.Canny(mask_bin, 60, 120)  # edges for "contour"
    edges_rgb = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)

    ok = False
    feedback = "No person detected"
    color = (255, 0, 0)  # RGB red

    if res.pose_landmarks:
        lm = res.pose_landmarks.landmark

        def pt(idx):
            return (lm[idx].x * w, lm[idx].y * h)

        P = mp_pose.PoseLandmark
        pts = {
            "l_sho": pt(P.LEFT_SHOULDER.value),
            "l_elb": pt(P.LEFT_ELBOW.value),
            "l_wri": pt(P.LEFT_WRIST.value),
            "r_sho": pt(P.RIGHT_SHOULDER.value),
            "r_elb": pt(P.RIGHT_ELBOW.value),
            "r_wri": pt(P.RIGHT_WRIST.value),
            "l_hip": pt(P.LEFT_HIP.value),
            "l_knee": pt(P.LEFT_KNEE.value),
            "l_ank": pt(P.LEFT_ANKLE.value),
            "r_hip": pt(P.RIGHT_HIP.value),
            "r_knee": pt(P.RIGHT_KNEE.value),
            "r_ank": pt(P.RIGHT_ANKLE.value),
        }

        ok, feedback = pose_rules(pose_name, pts)
        color = (0, 255, 0) if ok else (255, 0, 0)  # RGB green/red

        # draw skeleton
        mp_draw.draw_landmarks(
            img,
            res.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_draw.DrawingSpec(color=color, thickness=2, circle_radius=2),
            connection_drawing_spec=mp_draw.DrawingSpec(color=color, thickness=2),
        )

        # apply contour overlay (edges)
        # Ù†Ø¬Ù…Ø¹ edges Ø¹Ù„Ù‰ Ø§Ù„ØµÙˆØ±Ø© Ø¨Ù„ÙˆÙ† Ø§Ù„Ø­Ø§Ù„Ø©
        overlay = img.copy()
        overlay[edges > 0] = color  # Ø±Ø³Ù… ÙƒÙˆÙ†ØªÙˆØ± ØªÙ‚Ø±ÙŠØ¨ÙŠ
        img = cv2.addWeighted(img, 1.0, overlay, 0.6, 0)

        # status text
        txt = "âœ… Correct" if ok else "âŒ Fix posture"
        cv2.putText(img, f"{pose_name} | {txt}", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1.0, color, 3)
        cv2.putText(img, feedback, (20, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (255,255,255), 2)

    return img

demo = gr.Interface(
    fn=process,
    inputs=[
        gr.Image(sources=["webcam"], streaming=True, type="numpy", label="Camera"),
        gr.Dropdown(POSES, value="WarriorII", label="Choose Pose"),
    ],
    outputs=gr.Image(type="numpy", label="Result"),
    live=True,
    title="Yoga Pose Detection (Green=Correct, Red=Incorrect)",
    description="ÙŠÙØªØ­ Ø§Ù„ÙƒØ§Ù…ÙŠØ±Ø§ Ù…Ù† Ø§Ù„Ù…ØªØµÙØ­ØŒ ÙˆÙŠØ¹Ù…Ù„ Skeleton + Contour ØªÙ‚Ø±ÙŠØ¨Ù‹Ø§. Ø£Ø®Ø¶Ø± Ù„Ùˆ Ø§Ù„ÙˆØ¶Ø¹ÙŠØ© Ø¶Ù…Ù† Ø§Ù„Ù‚ÙˆØ§Ø¹Ø¯ØŒ Ø£Ø­Ù…Ø± Ù„Ùˆ ØºÙ„Ø·."
)

demo.launch(share=True)
```

```python
%%writefile /content/app.py
from flask import Flask, request, render_template, redirect, url_for, session
import torch
import csv
import os
import glob
import shutil
from datetime import date
import sqlite3
from werkzeug.security import generate_password_hash, check_password_hash
from pyngrok import ngrok

# ================== CONFIG ==================
MODEL_PATH = "/content/last (2).pt"      # Ø¹Ø¯Ù‘Ù„ Ù„Ùˆ Ø§Ø³Ù… Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„ Ù…Ø®ØªÙ„Ù
CSV_PATH   = "/content/caloris.csv"     # Ø¹Ø¯Ù‘Ù„ Ù„Ùˆ Ø§Ø³Ù… Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª Ù…Ø®ØªÙ„Ù
DB_PATH    = "/content/app.db"

NGROK_TOKEN = "35c97fFqjssPXEyCdZm46xBofrg_2Fxk6xyfyFYnLgaGUe8qb"  # ğŸ‘ˆ Ø­Ø· ØªÙˆÙƒÙ† Ø¬Ø¯ÙŠØ¯

# ================== APP ==================
app = Flask(__name__)
app.secret_key = "change-this-secret"

# ================== NORMALIZE LABELS ==================
def norm_label(s: str) -> str:
    s = (s or "").strip().lower()
    s = s.replace("_", " ")
    s = " ".join(s.split())
    return s

# ================== DATABASE ==================
def db_conn():
    return sqlite3.connect(DB_PATH)

def init_db():
    con = db_conn()
    cur = con.cursor()

    cur.execute(
        "CREATE TABLE IF NOT EXISTS users ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "name TEXT NOT NULL,"
        "email TEXT NOT NULL UNIQUE,"
        "password_hash TEXT NOT NULL"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS profiles ("
        "user_id INTEGER PRIMARY KEY,"
        "sex TEXT,"
        "age INTEGER,"
        "height_cm REAL,"
        "weight_kg REAL,"
        "activity TEXT,"
        "goal TEXT,"
        "daily_goal REAL,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    # âœ… workouts
    cur.execute(
        "CREATE TABLE IF NOT EXISTS workouts ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "user_id INTEGER NOT NULL,"
        "day TEXT NOT NULL,"
        "workout_type TEXT NOT NULL,"
        "minutes REAL NOT NULL,"
        "calories REAL NOT NULL,"
        "created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    con.commit()
    con.close()

init_db()

def get_user_by_email(email: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT id, name, email, password_hash FROM users WHERE email=?", (email,))
    row = cur.fetchone()
    con.close()
    return row

def create_user(name: str, email: str, password: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO users (name,email,password_hash) VALUES (?,?,?)",
        (name, email, generate_password_hash(password))
    )
    con.commit()
    uid = cur.lastrowid
    con.close()
    return uid

def upsert_profile(uid: int, sex: str, age: int, height: float, weight: float, activity: str, goal: str, daily: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO profiles VALUES (?,?,?,?,?,?,?,?) "
        "ON CONFLICT(user_id) DO UPDATE SET "
        "sex=excluded.sex, age=excluded.age, height_cm=excluded.height_cm, "
        "weight_kg=excluded.weight_kg, activity=excluded.activity, "
        "goal=excluded.goal, daily_goal=excluded.daily_goal",
        (uid, sex, age, height, weight, activity, goal, daily)
    )
    con.commit()
    con.close()

def get_daily_goal(uid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT daily_goal FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return row[0] if row else None

# ----- workouts db -----
def add_workout(uid: int, day: str, workout_type: str, minutes: float, calories: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO workouts (user_id, day, workout_type, minutes, calories) VALUES (?,?,?,?,?)",
        (uid, day, workout_type, minutes, calories)
    )
    con.commit()
    con.close()

def get_workouts(uid: int, day: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT workout_type, minutes, calories FROM workouts WHERE user_id=? AND day=? ORDER BY id DESC",
        (uid, day)
    )
    rows = cur.fetchall()
    con.close()
    return rows

def get_workouts_total(uid: int, day: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT COALESCE(SUM(calories),0) FROM workouts WHERE user_id=? AND day=?",
        (uid, day)
    )
    total = cur.fetchone()[0]
    con.close()
    return float(total or 0.0)

# ================== MODEL ==================
# âš ï¸ YOLOv5 via torch.hub may need ultralytics package (installed above)
model = torch.hub.load(
    "ultralytics/yolov5",
    "custom",
    path=MODEL_PATH,
    trust_repo=True
)

# ================== LOAD CALORIES CSV ==================
calorie_data = {}   # normalized -> kcal per 100g
label_display = {}  # normalized -> original label (from CSV)

with open(CSV_PATH, "r", encoding="cp1256", newline="") as f:
    reader = csv.reader(f)
    for row in reader:
        if not row or len(row) < 3:
            continue
        raw_name = row[0].strip()
        key = norm_label(raw_name)
        try:
            calorie_data[key] = float(row[2])
            label_display[key] = raw_name
        except:
            pass

# ================== CALC DAILY GOAL (BMR/TDEE) ==================
def calc_daily(sex, age, height, weight, activity, goal):
    sex = (sex or "").lower().strip()

    if sex == "male":
        bmr = 10*weight + 6.25*height - 5*age + 5
    else:
        bmr = 10*weight + 6.25*height - 5*age - 161

    factors = {
        "sedentary": 1.2,
        "light": 1.375,
        "moderate": 1.55,
        "active": 1.725,
        "very_active": 1.9
    }
    tdee = bmr * factors.get(activity, 1.2)

    if goal == "lose":
        tdee -= 500
    elif goal == "gain":
        tdee += 300

    tdee = max(tdee, 1200 if sex == "female" else 1500)
    return float(round(tdee, 0))

# ================== DAILY SESSION STATE ==================
def ensure_today():
    today = str(date.today())
    if session.get("day") != today:
        session["day"] = today
        session["meals"] = []
        session["consumed"] = 0.0

def require_login():
    return "user_id" in session

# ================== WORKOUT CAL ESTIMATE ==================
def estimate_calories(workout_type: str, minutes: float, weight_kg: float):
    mets = {
        "walking": 3.5,
        "running": 8.0,
        "cycling": 6.0,
        "gym": 5.0,
        "swimming": 7.0,
        "football": 7.0,
    }
    met = mets.get(workout_type, 4.0)
    return met * 3.5 * weight_kg / 200.0 * minutes

def get_user_weight(uid: int) -> float:
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT weight_kg FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return float(row[0]) if row and row[0] else 70.0

# ================== ROUTES ==================
@app.route("/")
def root():
    if not require_login():
        return redirect(url_for("login"))

    daily = get_daily_goal(session["user_id"])
    if daily is None:
        return redirect(url_for("onboarding"))

    session["daily_goal"] = float(daily)
    ensure_today()
    return redirect(url_for("dashboard"))

# ---------- AUTH ----------
@app.route("/signup", methods=["GET", "POST"])
def signup():
    if request.method == "POST":
        name = request.form.get("name", "").strip()
        email = request.form.get("email", "").strip().lower()
        password = request.form.get("password", "")

        if not name or not email or not password:
            return render_template("signup.html", error="Ø§Ù…Ù„Ø£ ÙƒÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª.")

        if get_user_by_email(email):
            return render_template("signup.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ Ù…ÙˆØ¬ÙˆØ¯ Ø¨Ø§Ù„ÙØ¹Ù„. Ø³Ø¬Ù„ Ø¯Ø®ÙˆÙ„.")

        uid = create_user(name, email, password)
        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = email

        return redirect(url_for("onboarding"))

    return render_template("signup.html")

@app.route("/login", methods=["GET", "POST"])
def login():
    if request.method == "POST":
        email = request.form.get("email", "").strip().lower()
        password = request.form.get("password", "")

        row = get_user_by_email(email)
        if not row:
            return render_template("login.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ ØºÙŠØ± ØµØ­ÙŠØ­.")

        uid, name, em, pw_hash = row
        if not check_password_hash(pw_hash, password):
            return render_template("login.html", error="ÙƒÙ„Ù…Ø© Ø§Ù„Ù…Ø±ÙˆØ± ØºÙŠØ± ØµØ­ÙŠØ­Ø©.")

        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = em

        return redirect(url_for("root"))

    return render_template("login.html")

@app.route("/logout")
def logout():
    session.clear()
    return redirect(url_for("login"))

# ---------- ONBOARDING ----------
@app.route("/onboarding", methods=["GET", "POST"])
def onboarding():
    if not require_login():
        return redirect(url_for("login"))

    if request.method == "POST":
        sex = request.form.get("sex")
        age = int(request.form.get("age"))
        height = float(request.form.get("height_cm"))
        weight = float(request.form.get("weight_kg"))
        activity = request.form.get("activity")
        goal = request.form.get("goal")

        daily = calc_daily(sex, age, height, weight, activity, goal)
        upsert_profile(session["user_id"], sex, age, height, weight, activity, goal, daily)

        session["daily_goal"] = daily
        ensure_today()
        return redirect(url_for("dashboard"))

    return render_template("onboarding.html")

# ---------- DASHBOARD ----------
@app.route("/dashboard")
def dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    daily = float(session["daily_goal"])
    consumed = float(session.get("consumed", 0.0))
    day = session.get("day")

    burned = get_workouts_total(session["user_id"], day)
    workouts = get_workouts(session["user_id"], day)

    net = max(0.0, round(consumed - burned, 2))
    remaining = max(0.0, round(daily - net, 2))
    over_by = max(0.0, round(net - daily, 2))

    return render_template(
        "dashboard.html",
        user_name=session.get("user_name"),
        daily_goal=round(daily, 0),
        consumed=round(consumed, 2),
        burned=round(burned, 2),
        net=net,
        remaining=remaining,
        over_by=over_by,
        meals=session.get("meals", []),
        workouts=workouts,
        day=day
    )

# ---------- MEALS ----------
@app.route("/start/<meal>")
def start_meal(meal):
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    meal = (meal or "").lower().strip()
    allowed = {"breakfast", "lunch", "dinner", "snack"}
    if meal not in allowed:
        meal = "breakfast"

    session["current_meal"] = meal
    return redirect(url_for("upload"))

@app.route("/upload", methods=["GET", "POST"])
def upload():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    meal = session.get("current_meal", "breakfast")

    if request.method == "POST":
        if "image" not in request.files or request.files["image"].filename == "":
            return render_template("upload.html", error="Ø§Ø±ÙØ¹ ØµÙˆØ±Ø© Ù…Ù† ÙØ¶Ù„Ùƒ.", meal_type=meal)

        img = request.files["image"]
        image_path = "/content/static/uploaded.jpg"
        img.save(image_path)

        results = model(image_path)
        df = results.pandas().xyxy[0]

        foods_raw = df["name"].unique().tolist() if len(df) else []
        foods = [norm_label(x) for x in foods_raw]

        annotated_url = None
        try:
            results.save()
            exp_dirs = sorted(glob.glob("runs/detect/exp*"), key=os.path.getmtime)
            latest = exp_dirs[-1] if exp_dirs else None
            if latest:
                imgs = glob.glob(os.path.join(latest, "*.jpg")) + glob.glob(os.path.join(latest, "*.png"))
                if imgs:
                    shutil.copy(imgs[0], "/content/static/annotated.jpg")
                    annotated_url = "/static/annotated.jpg"
        except:
            annotated_url = None

        res = []
        for k in foods:
            if k in calorie_data:
                res.append(f"{label_display.get(k, k)}: {calorie_data[k]} kcal / 100g")
            else:
                res.append(f"{label_display.get(k, k)}: Ù…Ø´ Ù…ÙˆØ¬ÙˆØ¯ ÙÙŠ Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª")

        return render_template(
            "results.html",
            foods=foods,
            res=res,
            annotated_url=annotated_url,
            daily_goal=session["daily_goal"],
            meal_type=meal
        )

    return render_template("upload.html", meal_type=meal)

@app.route("/calculate", methods=["POST"])
def calculate():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    foods = request.form.getlist("food")
    qtys  = request.form.getlist("quantity")

    total = 0.0
    details = []

    for f, q in zip(foods, qtys):
        key = norm_label(f)
        q = float(q)

        cal_per_100 = calorie_data.get(key, 0.0)
        cal = cal_per_100 * (q / 100.0)

        total += cal
        details.append({
            "food": label_display.get(key, key),
            "qty": q,
            "cal": round(cal, 2) if cal_per_100 > 0 else None
        })

    total = round(total, 2)

    session["meals"] = session.get("meals", [])
    session["meals"].append({
        "meal_type": session.get("current_meal", "meal"),
        "total": total,
        "details": details
    })

    session["consumed"] = float(session.get("consumed", 0.0)) + total
    return redirect(url_for("dashboard"))

# ---------- WORKOUTS ----------
@app.route("/workouts/add", methods=["POST"])
def workouts_add():
    if not require_login():
        return redirect(url_for("login"))
    ensure_today()

    wtype = request.form.get("workout_type", "walking")
    minutes = float(request.form.get("minutes", "0") or 0)
    day = session.get("day")

    weight = get_user_weight(session["user_id"])
    calories = float(round(estimate_calories(wtype, minutes, weight), 2))

    add_workout(session["user_id"], day, wtype, minutes, calories)
    return redirect(url_for("dashboard"))

# ================== RUN (for colab) ==================
def run_app():
    ngrok.kill()
    ngrok.set_auth_token(NGROK_TOKEN)
    public_url = ngrok.connect(5000)
    print("âœ… Ø§ÙØªØ­ Ø§Ù„Ù„ÙŠÙ†Ùƒ Ø¯Ù‡:", public_url)
    app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)

if __name__ == "__main__":
    run_app()
```

```python
%%writefile /content/app.py
from flask import Flask, request, render_template, redirect, url_for, session
import torch
import csv
import os
import glob
import shutil
from datetime import date
import sqlite3
from werkzeug.security import generate_password_hash, check_password_hash
from pyngrok import ngrok

# ================== CONFIG ==================
MODEL_PATH = "/content/last (2).pt"      # Ø¹Ø¯Ù‘Ù„ Ù„Ùˆ Ø§Ø³Ù… Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„ Ù…Ø®ØªÙ„Ù
CSV_PATH   = "/content/caloris.csv"     # Ø¹Ø¯Ù‘Ù„ Ù„Ùˆ Ø§Ø³Ù… Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª Ù…Ø®ØªÙ„Ù
DB_PATH    = "/content/app.db"

NGROK_TOKEN = "35c97fFqjssPXEyCdZm46xBofrg_2Fxk6xyfyFYnLgaGUe8qb"  # ğŸ‘ˆ Ø­Ø· ØªÙˆÙƒÙ† Ø¬Ø¯ÙŠØ¯

# ================== APP ==================
app = Flask(__name__)
app.secret_key = "change-this-secret"

# ================== NORMALIZE LABELS ==================
def norm_label(s: str) -> str:
    s = (s or "").strip().lower()
    s = s.replace("_", " ")
    s = " ".join(s.split())
    return s

# ================== DATABASE ==================
def db_conn():
    return sqlite3.connect(DB_PATH)

def init_db():
    con = db_conn()
    cur = con.cursor()

    cur.execute(
        "CREATE TABLE IF NOT EXISTS users ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "name TEXT NOT NULL,"
        "email TEXT NOT NULL UNIQUE,"
        "password_hash TEXT NOT NULL"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS profiles ("
        "user_id INTEGER PRIMARY KEY,"
        "sex TEXT,"
        "age INTEGER,"
        "height_cm REAL,"
        "weight_kg REAL,"
        "activity TEXT,"
        "goal TEXT,"
        "daily_goal REAL,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    # workouts
    cur.execute(
        "CREATE TABLE IF NOT EXISTS workouts ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "user_id INTEGER NOT NULL,"
        "day TEXT NOT NULL,"
        "workout_type TEXT NOT NULL,"
        "minutes REAL NOT NULL,"
        "calories REAL NOT NULL,"
        "created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    con.commit()
    con.close()

init_db()

def get_user_by_email(email: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT id, name, email, password_hash FROM users WHERE email=?", (email,))
    row = cur.fetchone()
    con.close()
    return row

def create_user(name: str, email: str, password: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO users (name,email,password_hash) VALUES (?,?,?)",
        (name, email, generate_password_hash(password))
    )
    con.commit()
    uid = cur.lastrowid
    con.close()
    return uid

def upsert_profile(uid: int, sex: str, age: int, height: float, weight: float, activity: str, goal: str, daily: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO profiles VALUES (?,?,?,?,?,?,?,?) "
        "ON CONFLICT(user_id) DO UPDATE SET "
        "sex=excluded.sex, age=excluded.age, height_cm=excluded.height_cm, "
        "weight_kg=excluded.weight_kg, activity=excluded.activity, "
        "goal=excluded.goal, daily_goal=excluded.daily_goal",
        (uid, sex, age, height, weight, activity, goal, daily)
    )
    con.commit()
    con.close()

def get_daily_goal(uid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT daily_goal FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return row[0] if row else None

# ----- workouts db -----
def add_workout(uid: int, day: str, workout_type: str, minutes: float, calories: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO workouts (user_id, day, workout_type, minutes, calories) VALUES (?,?,?,?,?)",
        (uid, day, workout_type, minutes, calories)
    )
    con.commit()
    con.close()

def get_workouts(uid: int, day: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT workout_type, minutes, calories FROM workouts WHERE user_id=? AND day=? ORDER BY id DESC",
        (uid, day)
    )
    rows = cur.fetchall()
    con.close()
    return rows

def get_workouts_total(uid: int, day: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT COALESCE(SUM(calories),0) FROM workouts WHERE user_id=? AND day=?",
        (uid, day)
    )
    total = cur.fetchone()[0]
    con.close()
    return float(total or 0.0)

# ================== MODEL ==================
model = torch.hub.load(
    "ultralytics/yolov5",
    "custom",
    path=MODEL_PATH,
    trust_repo=True
)

# ================== LOAD CALORIES CSV ==================
calorie_data = {}   # normalized -> kcal per 100g
label_display = {}  # normalized -> original label (from CSV)

with open(CSV_PATH, "r", encoding="cp1256", newline="") as f:
    reader = csv.reader(f)
    for row in reader:
        if not row or len(row) < 3:
            continue
        raw_name = row[0].strip()
        key = norm_label(raw_name)
        try:
            calorie_data[key] = float(row[2])
            label_display[key] = raw_name
        except:
            pass

# ================== CALC DAILY GOAL (BMR/TDEE) ==================
def calc_daily(sex, age, height, weight, activity, goal):
    sex = (sex or "").lower().strip()

    if sex == "male":
        bmr = 10*weight + 6.25*height - 5*age + 5
    else:
        bmr = 10*weight + 6.25*height - 5*age - 161

    factors = {
        "sedentary": 1.2,
        "light": 1.375,
        "moderate": 1.55,
        "active": 1.725,
        "very_active": 1.9
    }
    tdee = bmr * factors.get(activity, 1.2)

    if goal == "lose":
        tdee -= 500
    elif goal == "gain":
        tdee += 300

    tdee = max(tdee, 1200 if sex == "female" else 1500)
    return float(round(tdee, 0))

# ================== DAILY SESSION STATE ==================
def ensure_today():
    today = str(date.today())
    if session.get("day") != today:
        session["day"] = today
        session["meals"] = []
        session["consumed"] = 0.0

def require_login():
    return "user_id" in session

# ================== WORKOUT CAL ESTIMATE ==================
def estimate_calories(workout_type: str, minutes: float, weight_kg: float):
    mets = {
        # Cardio
        "walking": 3.5,
        "running": 8.0,
        "cycling": 6.0,
        "swimming": 7.0,
        "jump_rope": 10.0,
        "elliptical": 5.0,
        "stair_climber": 8.8,
        "rowing": 7.0,
        "hiit": 9.0,
        "football": 7.0,
        "basketball": 8.0,

        # Gym
        "weights_light": 3.0,
        "weights_moderate": 5.0,
        "weights_heavy": 6.0,
        "circuit_training": 8.0,
        "crossfit": 10.0,
    }
    met = mets.get(workout_type, 4.0)
    return met * 3.5 * weight_kg / 200.0 * minutes

def get_user_weight(uid: int) -> float:
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT weight_kg FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return float(row[0]) if row and row[0] else 70.0

# ================== ROUTES ==================
@app.route("/")
def root():
    if not require_login():
        return redirect(url_for("login"))

    daily = get_daily_goal(session["user_id"])
    if daily is None:
        return redirect(url_for("onboarding"))

    session["daily_goal"] = float(daily)
    ensure_today()
    return redirect(url_for("home"))  # âœ… ØµÙØ­Ø© Ø§Ù„Ø§Ø®ØªÙŠØ§Ø±

@app.route("/home")
def home():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))
    ensure_today()
    return render_template("home.html")

@app.route("/go/diet")
def go_diet():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("dashboard") + "?tab=food")

@app.route("/go/workout")
def go_workout():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("dashboard") + "?tab=workout")

# ---------- AUTH ----------
@app.route("/signup", methods=["GET", "POST"])
def signup():
    if request.method == "POST":
        name = request.form.get("name", "").strip()
        email = request.form.get("email", "").strip().lower()
        password = request.form.get("password", "")

        if not name or not email or not password:
            return render_template("signup.html", error="Ø§Ù…Ù„Ø£ ÙƒÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª.")

        if get_user_by_email(email):
            return render_template("signup.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ Ù…ÙˆØ¬ÙˆØ¯ Ø¨Ø§Ù„ÙØ¹Ù„. Ø³Ø¬Ù„ Ø¯Ø®ÙˆÙ„.")

        uid = create_user(name, email, password)
        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = email
        return redirect(url_for("onboarding"))

    return render_template("signup.html")

@app.route("/login", methods=["GET", "POST"])
def login():
    if request.method == "POST":
        email = request.form.get("email", "").strip().lower()
        password = request.form.get("password", "")

        row = get_user_by_email(email)
        if not row:
            return render_template("login.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ ØºÙŠØ± ØµØ­ÙŠØ­.")

        uid, name, em, pw_hash = row
        if not check_password_hash(pw_hash, password):
            return render_template("login.html", error="ÙƒÙ„Ù…Ø© Ø§Ù„Ù…Ø±ÙˆØ± ØºÙŠØ± ØµØ­ÙŠØ­Ø©.")

        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = em
        return redirect(url_for("root"))

    return render_template("login.html")

@app.route("/logout")
def logout():
    session.clear()
    return redirect(url_for("login"))

# ---------- ONBOARDING ----------
@app.route("/onboarding", methods=["GET", "POST"])
def onboarding():
    if not require_login():
        return redirect(url_for("login"))

    if request.method == "POST":
        sex = request.form.get("sex")
        age = int(request.form.get("age"))
        height = float(request.form.get("height_cm"))
        weight = float(request.form.get("weight_kg"))
        activity = request.form.get("activity")
        goal = request.form.get("goal")

        daily = calc_daily(sex, age, height, weight, activity, goal)
        upsert_profile(session["user_id"], sex, age, height, weight, activity, goal, daily)

        session["daily_goal"] = daily
        ensure_today()
        return redirect(url_for("home"))

    return render_template("onboarding.html")

# ---------- DASHBOARD ----------
@app.route("/dashboard")
def dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    daily = float(session["daily_goal"])
    consumed = float(session.get("consumed", 0.0))
    day = session.get("day")

    burned = get_workouts_total(session["user_id"], day)
    workouts = get_workouts(session["user_id"], day)

    net = max(0.0, round(consumed - burned, 2))
    remaining = max(0.0, round(daily - net, 2))
    over_by = max(0.0, round(net - daily, 2))

    return render_template(
        "dashboard.html",
        user_name=session.get("user_name"),
        daily_goal=round(daily, 0),
        consumed=round(consumed, 2),
        burned=round(burned, 2),
        net=net,
        remaining=remaining,
        over_by=over_by,
        meals=session.get("meals", []),
        workouts=workouts,
        day=day
    )

# ---------- MEALS ----------
@app.route("/start/<meal>")
def start_meal(meal):
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    meal = (meal or "").lower().strip()
    allowed = {"breakfast", "lunch", "dinner", "snack"}
    if meal not in allowed:
        meal = "breakfast"

    session["current_meal"] = meal
    return redirect(url_for("upload"))

@app.route("/upload", methods=["GET", "POST"])
def upload():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    meal = session.get("current_meal", "breakfast")

    if request.method == "POST":
        if "image" not in request.files or request.files["image"].filename == "":
            return render_template("upload.html", error="Ø§Ø±ÙØ¹ ØµÙˆØ±Ø© Ù…Ù† ÙØ¶Ù„Ùƒ.", meal_type=meal)

        img = request.files["image"]
        image_path = "/content/static/uploaded.jpg"
        img.save(image_path)

        results = model(image_path)
        df = results.pandas().xyxy[0]

        foods_raw = df["name"].unique().tolist() if len(df) else []
        foods = [norm_label(x) for x in foods_raw]

        annotated_url = None
        try:
            results.save()
            exp_dirs = sorted(glob.glob("runs/detect/exp*"), key=os.path.getmtime)
            latest = exp_dirs[-1] if exp_dirs else None
            if latest:
                imgs = glob.glob(os.path.join(latest, "*.jpg")) + glob.glob(os.path.join(latest, "*.png"))
                if imgs:
                    shutil.copy(imgs[0], "/content/static/annotated.jpg")
                    annotated_url = "/static/annotated.jpg"
        except:
            annotated_url = None

        res = []
        for k in foods:
            if k in calorie_data:
                res.append(f"{label_display.get(k, k)}: {calorie_data[k]} kcal / 100g")
            else:
                res.append(f"{label_display.get(k, k)}: Ù…Ø´ Ù…ÙˆØ¬ÙˆØ¯ ÙÙŠ Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª")

        return render_template(
            "results.html",
            foods=foods,
            res=res,
            annotated_url=annotated_url,
            daily_goal=session["daily_goal"],
            meal_type=meal
        )

    return render_template("upload.html", meal_type=meal)

@app.route("/calculate", methods=["POST"])
def calculate():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    foods = request.form.getlist("food")
    qtys  = request.form.getlist("quantity")

    total = 0.0
    details = []

    for f, q in zip(foods, qtys):
        key = norm_label(f)
        q = float(q)

        cal_per_100 = calorie_data.get(key, 0.0)
        cal = cal_per_100 * (q / 100.0)

        total += cal
        details.append({
            "food": label_display.get(key, key),
            "qty": q,
            "cal": round(cal, 2) if cal_per_100 > 0 else None
        })

    total = round(total, 2)

    session["meals"] = session.get("meals", [])
    session["meals"].append({
        "meal_type": session.get("current_meal", "meal"),
        "total": total,
        "details": details
    })

    session["consumed"] = float(session.get("consumed", 0.0)) + total
    return redirect(url_for("dashboard"))

# ---------- WORKOUTS ----------
@app.route("/workouts/add", methods=["POST"])
def workouts_add():
    if not require_login():
        return redirect(url_for("login"))
    ensure_today()

    wtype = request.form.get("workout_type", "walking")
    minutes = float(request.form.get("minutes", "0") or 0)
    day = session.get("day")

    weight = get_user_weight(session["user_id"])
    calories = float(round(estimate_calories(wtype, minutes, weight), 2))

    add_workout(session["user_id"], day, wtype, minutes, calories)
    return redirect(url_for("dashboard") + "?tab=workout")

# ================== RUN (colab) ==================
def run_app():
    ngrok.kill()
    ngrok.set_auth_token(NGROK_TOKEN)
    public_url = ngrok.connect(5000)
    print("âœ… Ø§ÙØªØ­ Ø§Ù„Ù„ÙŠÙ†Ùƒ Ø¯Ù‡:", public_url)
    app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)

if __name__ == "__main__":
    run_app()
```

```python
%%writefile /content/app.py
from flask import Flask, request, render_template, redirect, url_for, session
import torch
import csv
import os
import glob
import shutil
from datetime import date
import sqlite3
from werkzeug.security import generate_password_hash, check_password_hash
from pyngrok import ngrok

# ================== CONFIG ==================
MODEL_PATH = "/content/last (2).pt"
CSV_PATH   = "/content/caloris.csv"
DB_PATH    = "/content/app.db"

NGROK_TOKEN = "35c97fFqjssPXEyCdZm46xBofrg_2Fxk6xyfyFYnLgaGUe8qb"  # ğŸ‘ˆ Ø­Ø· ØªÙˆÙƒÙ† Ø¬Ø¯ÙŠØ¯

# ================== APP ==================
app = Flask(__name__)
app.secret_key = "change-this-secret"

# ================== NORMALIZE LABELS ==================
def norm_label(s: str) -> str:
    s = (s or "").strip().lower()
    s = s.replace("_", " ")
    s = " ".join(s.split())
    return s

# ================== DATABASE ==================
def db_conn():
    return sqlite3.connect(DB_PATH)

def init_db():
    con = db_conn()
    cur = con.cursor()

    cur.execute(
        "CREATE TABLE IF NOT EXISTS users ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "name TEXT NOT NULL,"
        "email TEXT NOT NULL UNIQUE,"
        "password_hash TEXT NOT NULL"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS profiles ("
        "user_id INTEGER PRIMARY KEY,"
        "sex TEXT,"
        "age INTEGER,"
        "height_cm REAL,"
        "weight_kg REAL,"
        "activity TEXT,"
        "goal TEXT,"
        "daily_goal REAL,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS workouts ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "user_id INTEGER NOT NULL,"
        "day TEXT NOT NULL,"
        "workout_type TEXT NOT NULL,"
        "minutes REAL NOT NULL,"
        "calories REAL NOT NULL,"
        "created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    con.commit()
    con.close()

init_db()

def get_user_by_email(email: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT id, name, email, password_hash FROM users WHERE email=?", (email,))
    row = cur.fetchone()
    con.close()
    return row

def create_user(name: str, email: str, password: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO users (name,email,password_hash) VALUES (?,?,?)",
        (name, email, generate_password_hash(password))
    )
    con.commit()
    uid = cur.lastrowid
    con.close()
    return uid

def upsert_profile(uid: int, sex: str, age: int, height: float, weight: float, activity: str, goal: str, daily: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO profiles VALUES (?,?,?,?,?,?,?,?) "
        "ON CONFLICT(user_id) DO UPDATE SET "
        "sex=excluded.sex, age=excluded.age, height_cm=excluded.height_cm, "
        "weight_kg=excluded.weight_kg, activity=excluded.activity, "
        "goal=excluded.goal, daily_goal=excluded.daily_goal",
        (uid, sex, age, height, weight, activity, goal, daily)
    )
    con.commit()
    con.close()

def get_daily_goal(uid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT daily_goal FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return row[0] if row else None

# ----- workouts db -----
def add_workout(uid: int, day: str, workout_type: str, minutes: float, calories: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO workouts (user_id, day, workout_type, minutes, calories) VALUES (?,?,?,?,?)",
        (uid, day, workout_type, minutes, calories)
    )
    con.commit()
    con.close()

def get_workouts(uid: int, day: str):
    # âœ… Ø±Ø¬Ù‘Ø¹ id ÙƒÙ…Ø§Ù† Ø¹Ù„Ø´Ø§Ù† Ø§Ù„Ø­Ø°Ù
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT workout_type, minutes, calories, id FROM workouts WHERE user_id=? AND day=? ORDER BY id DESC",
        (uid, day)
    )
    rows = cur.fetchall()
    con.close()
    return rows

def get_workouts_total(uid: int, day: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT COALESCE(SUM(calories),0) FROM workouts WHERE user_id=? AND day=?",
        (uid, day)
    )
    total = cur.fetchone()[0]
    con.close()
    return float(total or 0.0)

def delete_workout(uid: int, wid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("DELETE FROM workouts WHERE id=? AND user_id=?", (wid, uid))
    con.commit()
    con.close()

# ================== MODEL ==================
model = torch.hub.load(
    "ultralytics/yolov5",
    "custom",
    path=MODEL_PATH,
    trust_repo=True
)

# ================== LOAD CALORIES CSV ==================
calorie_data = {}
label_display = {}

with open(CSV_PATH, "r", encoding="cp1256", newline="") as f:
    reader = csv.reader(f)
    for row in reader:
        if not row or len(row) < 3:
            continue
        raw_name = row[0].strip()
        key = norm_label(raw_name)
        try:
            calorie_data[key] = float(row[2])
            label_display[key] = raw_name
        except:
            pass

# ================== CALC DAILY GOAL ==================
def calc_daily(sex, age, height, weight, activity, goal):
    sex = (sex or "").lower().strip()
    if sex == "male":
        bmr = 10*weight + 6.25*height - 5*age + 5
    else:
        bmr = 10*weight + 6.25*height - 5*age - 161

    factors = {"sedentary":1.2,"light":1.375,"moderate":1.55,"active":1.725,"very_active":1.9}
    tdee = bmr * factors.get(activity, 1.2)

    if goal == "lose":
        tdee -= 500
    elif goal == "gain":
        tdee += 300

    tdee = max(tdee, 1200 if sex == "female" else 1500)
    return float(round(tdee, 0))

# ================== SESSION ==================
def ensure_today():
    today = str(date.today())
    if session.get("day") != today:
        session["day"] = today
        session["meals"] = []
        session["consumed"] = 0.0

def require_login():
    return "user_id" in session

# ================== WORKOUT CAL ESTIMATE ==================
def estimate_calories(workout_type: str, minutes: float, weight_kg: float):
    mets = {
        # Cardio
        "walking": 3.5, "running": 8.0, "cycling": 6.0, "swimming": 7.0,
        "jump_rope": 10.0, "elliptical": 5.0, "stair_climber": 8.8,
        "rowing": 7.0, "hiit": 9.0, "football": 7.0, "basketball": 8.0,
        # Gym
        "weights_light": 3.0, "weights_moderate": 5.0, "weights_heavy": 6.0,
        "circuit_training": 8.0, "crossfit": 10.0,
    }
    met = mets.get(workout_type, 4.0)
    return met * 3.5 * weight_kg / 200.0 * minutes

def get_user_weight(uid: int) -> float:
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT weight_kg FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return float(row[0]) if row and row[0] else 70.0

# ================== ROUTES ==================
@app.route("/")
def root():
    if not require_login():
        return redirect(url_for("login"))
    daily = get_daily_goal(session["user_id"])
    if daily is None:
        return redirect(url_for("onboarding"))
    session["daily_goal"] = float(daily)
    ensure_today()
    return redirect(url_for("home"))

@app.route("/home")
def home():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))
    ensure_today()
    return render_template("home.html")

@app.route("/go/diet")
def go_diet():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("diet_dashboard"))

@app.route("/go/workout")
def go_workout():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("workout_dashboard"))

# âœ… Ø¥Ø¹Ø§Ø¯Ø© Ø­Ø³Ø§Ø¨ Ø§Ù„Ù‡Ø¯Ù (ÙŠØ±Ø¬Ø¹Ùƒ Onboarding)
@app.route("/recalc")
def recalc():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("onboarding"))

# ---------- AUTH ----------
@app.route("/signup", methods=["GET","POST"])
def signup():
    if request.method == "POST":
        name = request.form.get("name","").strip()
        email = request.form.get("email","").strip().lower()
        password = request.form.get("password","")

        if not name or not email or not password:
            return render_template("signup.html", error="Ø§Ù…Ù„Ø£ ÙƒÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª.")
        if get_user_by_email(email):
            return render_template("signup.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ Ù…ÙˆØ¬ÙˆØ¯ Ø¨Ø§Ù„ÙØ¹Ù„. Ø³Ø¬Ù„ Ø¯Ø®ÙˆÙ„.")

        uid = create_user(name, email, password)
        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = email
        return redirect(url_for("onboarding"))

    return render_template("signup.html")

@app.route("/login", methods=["GET","POST"])
def login():
    if request.method == "POST":
        email = request.form.get("email","").strip().lower()
        password = request.form.get("password","")

        row = get_user_by_email(email)
        if not row:
            return render_template("login.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ ØºÙŠØ± ØµØ­ÙŠØ­.")
        uid, name, em, pw_hash = row
        if not check_password_hash(pw_hash, password):
            return render_template("login.html", error="ÙƒÙ„Ù…Ø© Ø§Ù„Ù…Ø±ÙˆØ± ØºÙŠØ± ØµØ­ÙŠØ­Ø©.")

        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = em
        return redirect(url_for("root"))

    return render_template("login.html")

@app.route("/logout")
def logout():
    session.clear()
    return redirect(url_for("login"))

# ---------- ONBOARDING ----------
@app.route("/onboarding", methods=["GET","POST"])
def onboarding():
    if not require_login():
        return redirect(url_for("login"))

    if request.method == "POST":
        sex = request.form.get("sex")
        age = int(request.form.get("age"))
        height = float(request.form.get("height_cm"))
        weight = float(request.form.get("weight_kg"))
        activity = request.form.get("activity")
        goal = request.form.get("goal")

        daily = calc_daily(sex, age, height, weight, activity, goal)
        upsert_profile(session["user_id"], sex, age, height, weight, activity, goal, daily)

        session["daily_goal"] = daily
        ensure_today()
        return redirect(url_for("home"))

    return render_template("onboarding.html")

# ---------- DIET DASHBOARD ----------
@app.route("/diet")
def diet_dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    daily = float(session["daily_goal"])
    consumed = float(session.get("consumed", 0.0))

    remaining = max(0.0, round(daily - consumed, 2))
    over_by = max(0.0, round(consumed - daily, 2))

    return render_template(
        "diet_dashboard.html",
        user_name=session.get("user_name"),
        daily_goal=round(daily, 0),
        consumed=round(consumed, 2),
        remaining=remaining,
        over_by=over_by,
        meals=session.get("meals", []),
        day=session.get("day")
    )

# âœ… Ø­Ø°Ù ÙˆØ¬Ø¨Ø© Ø¨Ø§Ù„Ù€ index ÙÙŠ session
@app.route("/meals/delete/<int:idx>", methods=["POST"])
def meals_delete(idx):
    if not require_login():
        return redirect(url_for("login"))
    ensure_today()

    meals = session.get("meals", [])
    if 0 <= idx < len(meals):
        removed = meals.pop(idx)
        session["meals"] = meals

        # Ø±Ø¬Ù‘Ø¹ consumed
        removed_total = float(removed.get("total", 0.0) or 0.0)
        session["consumed"] = max(0.0, float(session.get("consumed", 0.0)) - removed_total)

    return redirect(url_for("diet_dashboard"))

# ---------- WORKOUT DASHBOARD ----------
@app.route("/workout")
def workout_dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    day = session.get("day")

    burned = get_workouts_total(session["user_id"], day)
    workouts = get_workouts(session["user_id"], day)

    return render_template(
        "workout_dashboard.html",
        user_name=session.get("user_name"),
        burned=round(burned, 2),
        workouts=workouts,
        day=day
    )

@app.route("/workouts/add", methods=["POST"])
def workouts_add():
    if not require_login():
        return redirect(url_for("login"))
    ensure_today()

    wtype = request.form.get("workout_type","walking")
    minutes = float(request.form.get("minutes","0") or 0)
    day = session.get("day")

    weight = get_user_weight(session["user_id"])
    calories = float(round(estimate_calories(wtype, minutes, weight), 2))

    add_workout(session["user_id"], day, wtype, minutes, calories)
    return redirect(url_for("workout_dashboard"))

# âœ… Ø­Ø°Ù ØªÙ…Ø±ÙŠÙ† Ù…Ù† DB Ø¨Ø§Ù„Ù€ id
@app.route("/workouts/delete/<int:wid>", methods=["POST"])
def workouts_delete(wid):
    if not require_login():
        return redirect(url_for("login"))
    delete_workout(session["user_id"], wid)
    return redirect(url_for("workout_dashboard"))

# ---------- MEALS FLOW ----------
@app.route("/start/<meal>")
def start_meal(meal):
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    meal = (meal or "").lower().strip()
    allowed = {"breakfast","lunch","dinner","snack"}
    if meal not in allowed:
        meal = "breakfast"

    session["current_meal"] = meal
    return redirect(url_for("upload"))

@app.route("/upload", methods=["GET","POST"])
def upload():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    meal = session.get("current_meal","breakfast")

    if request.method == "POST":
        if "image" not in request.files or request.files["image"].filename == "":
            return render_template("upload.html", error="Ø§Ø±ÙØ¹ ØµÙˆØ±Ø© Ù…Ù† ÙØ¶Ù„Ùƒ.", meal_type=meal)

        img = request.files["image"]
        image_path = "/content/static/uploaded.jpg"
        img.save(image_path)

        results = model(image_path)
        df = results.pandas().xyxy[0]

        foods_raw = df["name"].unique().tolist() if len(df) else []
        foods = [norm_label(x) for x in foods_raw]

        annotated_url = None
        try:
            results.save()
            exp_dirs = sorted(glob.glob("runs/detect/exp*"), key=os.path.getmtime)
            latest = exp_dirs[-1] if exp_dirs else None
            if latest:
                imgs = glob.glob(os.path.join(latest, "*.jpg")) + glob.glob(os.path.join(latest, "*.png"))
                if imgs:
                    shutil.copy(imgs[0], "/content/static/annotated.jpg")
                    annotated_url = "/static/annotated.jpg"
        except:
            annotated_url = None

        res = []
        for k in foods:
            if k in calorie_data:
                res.append(f"{label_display.get(k, k)}: {calorie_data[k]} kcal / 100g")
            else:
                res.append(f"{label_display.get(k, k)}: Ù…Ø´ Ù…ÙˆØ¬ÙˆØ¯ ÙÙŠ Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª")

        return render_template(
            "results.html",
            foods=foods,
            res=res,
            annotated_url=annotated_url,
            daily_goal=session["daily_goal"],
            meal_type=meal
        )

    return render_template("upload.html", meal_type=meal)

@app.route("/calculate", methods=["POST"])
def calculate():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    foods = request.form.getlist("food")
    qtys  = request.form.getlist("quantity")

    total = 0.0
    details = []

    for f, q in zip(foods, qtys):
        key = norm_label(f)
        q = float(q)

        cal_per_100 = calorie_data.get(key, 0.0)
        cal = cal_per_100 * (q / 100.0)

        total += cal
        details.append({
            "food": label_display.get(key, key),
            "qty": q,
            "cal": round(cal, 2) if cal_per_100 > 0 else None
        })

    total = round(total, 2)

    session["meals"] = session.get("meals", [])
    session["meals"].append({
        "meal_type": session.get("current_meal","meal"),
        "total": total,
        "details": details
    })

    session["consumed"] = float(session.get("consumed", 0.0)) + total
    return redirect(url_for("diet_dashboard"))

# ================== RUN (colab) ==================
def run_app():
    ngrok.kill()
    ngrok.set_auth_token(NGROK_TOKEN)
    public_url = ngrok.connect(5000)
    print("âœ… Ø§ÙØªØ­ Ø§Ù„Ù„ÙŠÙ†Ùƒ Ø¯Ù‡:", public_url)
    app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)

if __name__ == "__main__":
    run_app()
```

```python
%%writefile /content/app.py
from flask import Flask, request, render_template, redirect, url_for, session
import torch
import csv
import os
import glob
import shutil
from datetime import date
import sqlite3
from werkzeug.security import generate_password_hash, check_password_hash
from pyngrok import ngrok

# ================== CONFIG ==================
MODEL_PATH = "/content/last (2).pt"      # Ø¹Ø¯Ù‘Ù„ Ù„Ùˆ Ø§Ø³Ù… Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„ Ù…Ø®ØªÙ„Ù
CSV_PATH   = "/content/caloris.csv"     # Ø¹Ø¯Ù‘Ù„ Ù„Ùˆ Ø§Ø³Ù… Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª Ù…Ø®ØªÙ„Ù
DB_PATH    = "/content/app.db"

NGROK_TOKEN = "35c97fFqjssPXEyCdZm46xBofrg_2Fxk6xyfyFYnLgaGUe8qb"  # ğŸ‘ˆ Ø­Ø· ØªÙˆÙƒÙ† Ø¬Ø¯ÙŠØ¯

# ================== APP ==================
app = Flask(__name__)
app.secret_key = "change-this-secret"

# ================== NORMALIZE LABELS ==================
def norm_label(s: str) -> str:
    s = (s or "").strip().lower()
    s = s.replace("_", " ")
    s = " ".join(s.split())
    return s

# ================== DATABASE ==================
def db_conn():
    return sqlite3.connect(DB_PATH)

def init_db():
    con = db_conn()
    cur = con.cursor()

    cur.execute(
        "CREATE TABLE IF NOT EXISTS users ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "name TEXT NOT NULL,"
        "email TEXT NOT NULL UNIQUE,"
        "password_hash TEXT NOT NULL"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS profiles ("
        "user_id INTEGER PRIMARY KEY,"
        "sex TEXT,"
        "age INTEGER,"
        "height_cm REAL,"
        "weight_kg REAL,"
        "activity TEXT,"
        "goal TEXT,"
        "daily_goal REAL,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS workouts ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "user_id INTEGER NOT NULL,"
        "day TEXT NOT NULL,"
        "workout_type TEXT NOT NULL,"
        "minutes REAL NOT NULL,"
        "calories REAL NOT NULL,"
        "created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    con.commit()
    con.close()

init_db()

def get_user_by_email(email: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT id, name, email, password_hash FROM users WHERE email=?", (email,))
    row = cur.fetchone()
    con.close()
    return row

def create_user(name: str, email: str, password: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO users (name,email,password_hash) VALUES (?,?,?)",
        (name, email, generate_password_hash(password))
    )
    con.commit()
    uid = cur.lastrowid
    con.close()
    return uid

def upsert_profile(uid: int, sex: str, age: int, height: float, weight: float, activity: str, goal: str, daily: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO profiles VALUES (?,?,?,?,?,?,?,?) "
        "ON CONFLICT(user_id) DO UPDATE SET "
        "sex=excluded.sex, age=excluded.age, height_cm=excluded.height_cm, "
        "weight_kg=excluded.weight_kg, activity=excluded.activity, "
        "goal=excluded.goal, daily_goal=excluded.daily_goal",
        (uid, sex, age, height, weight, activity, goal, daily)
    )
    con.commit()
    con.close()

def get_daily_goal(uid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT daily_goal FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return row[0] if row else None

def get_user_weight(uid: int) -> float:
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT weight_kg FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return float(row[0]) if row and row[0] else 70.0

# ----- workouts db -----
def add_workout(uid: int, day: str, workout_type: str, minutes: float, calories: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO workouts (user_id, day, workout_type, minutes, calories) VALUES (?,?,?,?,?)",
        (uid, day, workout_type, minutes, calories)
    )
    con.commit()
    con.close()

def get_workouts(uid: int, day: str):
    # âœ… Ø±Ø¬Ù‘Ø¹ id ÙƒÙ…Ø§Ù† Ø¹Ù„Ø´Ø§Ù† Ø§Ù„Ø­Ø°Ù
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT workout_type, minutes, calories, id FROM workouts WHERE user_id=? AND day=? ORDER BY id DESC",
        (uid, day)
    )
    rows = cur.fetchall()
    con.close()
    return rows

def get_workouts_total(uid: int, day: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT COALESCE(SUM(calories),0) FROM workouts WHERE user_id=? AND day=?",
        (uid, day)
    )
    total = cur.fetchone()[0]
    con.close()
    return float(total or 0.0)

def delete_workout(uid: int, wid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("DELETE FROM workouts WHERE id=? AND user_id=?", (wid, uid))
    con.commit()
    con.close()

# ================== MODEL ==================
model = torch.hub.load(
    "ultralytics/yolov5",
    "custom",
    path=MODEL_PATH,
    trust_repo=True
)

# ================== LOAD CALORIES CSV ==================
calorie_data = {}
label_display = {}

with open(CSV_PATH, "r", encoding="cp1256", newline="") as f:
    reader = csv.reader(f)
    for row in reader:
        if not row or len(row) < 3:
            continue
        raw_name = row[0].strip()
        key = norm_label(raw_name)
        try:
            calorie_data[key] = float(row[2])
            label_display[key] = raw_name
        except:
            pass

# ================== CALC DAILY GOAL ==================
def calc_daily(sex, age, height, weight, activity, goal):
    sex = (sex or "").lower().strip()
    if sex == "male":
        bmr = 10*weight + 6.25*height - 5*age + 5
    else:
        bmr = 10*weight + 6.25*height - 5*age - 161

    factors = {"sedentary":1.2,"light":1.375,"moderate":1.55,"active":1.725,"very_active":1.9}
    tdee = bmr * factors.get(activity, 1.2)

    if goal == "lose":
        tdee -= 500
    elif goal == "gain":
        tdee += 300

    tdee = max(tdee, 1200 if sex == "female" else 1500)
    return float(round(tdee, 0))

# ================== SESSION ==================
def ensure_today():
    today = str(date.today())
    if session.get("day") != today:
        session["day"] = today
        session["meals"] = []
        session["consumed"] = 0.0

def require_login():
    return "user_id" in session

# ================== WORKOUT CAL ESTIMATE ==================
def estimate_calories(workout_type: str, minutes: float, weight_kg: float):
    mets = {
        # Cardio
        "walking": 3.5, "running": 8.0, "cycling": 6.0, "swimming": 7.0,
        "jump_rope": 10.0, "elliptical": 5.0, "stair_climber": 8.8,
        "rowing": 7.0, "hiit": 9.0, "football": 7.0, "basketball": 8.0,
        # Gym
        "weights_light": 3.0, "weights_moderate": 5.0, "weights_heavy": 6.0,
        "circuit_training": 8.0, "crossfit": 10.0,
    }
    met = mets.get(workout_type, 4.0)
    return met * 3.5 * weight_kg / 200.0 * minutes

# ================== ROUTES ==================
@app.route("/")
def root():
    if not require_login():
        return redirect(url_for("login"))
    daily = get_daily_goal(session["user_id"])
    if daily is None:
        return redirect(url_for("onboarding"))
    session["daily_goal"] = float(daily)
    ensure_today()
    return redirect(url_for("home"))

@app.route("/home")
def home():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))
    ensure_today()
    return render_template("home.html")

@app.route("/go/diet")
def go_diet():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("diet_dashboard"))

@app.route("/go/workout")
def go_workout():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("workout_dashboard"))

# âœ… Ø¥Ø¹Ø§Ø¯Ø© Ø­Ø³Ø§Ø¨ Ø§Ù„Ù‡Ø¯Ù (ÙŠØ±Ø¬Ø¹Ùƒ Onboarding)
@app.route("/recalc")
def recalc():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("onboarding"))

# ---------- AUTH ----------
@app.route("/signup", methods=["GET","POST"])
def signup():
    if request.method == "POST":
        name = request.form.get("name","").strip()
        email = request.form.get("email","").strip().lower()
        password = request.form.get("password","")

        if not name or not email or not password:
            return render_template("signup.html", error="Ø§Ù…Ù„Ø£ ÙƒÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª.")
        if get_user_by_email(email):
            return render_template("signup.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ Ù…ÙˆØ¬ÙˆØ¯ Ø¨Ø§Ù„ÙØ¹Ù„. Ø³Ø¬Ù„ Ø¯Ø®ÙˆÙ„.")

        uid = create_user(name, email, password)
        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = email
        return redirect(url_for("onboarding"))

    return render_template("signup.html")

@app.route("/login", methods=["GET","POST"])
def login():
    if request.method == "POST":
        email = request.form.get("email","").strip().lower()
        password = request.form.get("password","")

        row = get_user_by_email(email)
        if not row:
            return render_template("login.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ ØºÙŠØ± ØµØ­ÙŠØ­.")
        uid, name, em, pw_hash = row
        if not check_password_hash(pw_hash, password):
            return render_template("login.html", error="ÙƒÙ„Ù…Ø© Ø§Ù„Ù…Ø±ÙˆØ± ØºÙŠØ± ØµØ­ÙŠØ­Ø©.")

        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = em
        return redirect(url_for("root"))

    return render_template("login.html")

@app.route("/logout")
def logout():
    session.clear()
    return redirect(url_for("login"))

# ---------- ONBOARDING ----------
@app.route("/onboarding", methods=["GET","POST"])
def onboarding():
    if not require_login():
        return redirect(url_for("login"))

    if request.method == "POST":
        sex = request.form.get("sex")
        age = int(request.form.get("age"))
        height = float(request.form.get("height_cm"))
        weight = float(request.form.get("weight_kg"))
        activity = request.form.get("activity")
        goal = request.form.get("goal")

        daily = calc_daily(sex, age, height, weight, activity, goal)
        upsert_profile(session["user_id"], sex, age, height, weight, activity, goal, daily)

        session["daily_goal"] = daily
        ensure_today()
        return redirect(url_for("home"))

    return render_template("onboarding.html")

# ---------- DIET DASHBOARD ----------
@app.route("/diet")
def diet_dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    daily = float(session["daily_goal"])
    consumed = float(session.get("consumed", 0.0))

    remaining = max(0.0, round(daily - consumed, 2))
    over_by = max(0.0, round(consumed - daily, 2))

    return render_template(
        "diet_dashboard.html",
        user_name=session.get("user_name"),
        daily_goal=round(daily, 0),
        consumed=round(consumed, 2),
        remaining=remaining,
        over_by=over_by,
        meals=session.get("meals", []),
        day=session.get("day")
    )

# âœ… Ø­Ø°Ù ÙˆØ¬Ø¨Ø© Ø¨Ø§Ù„Ù€ index ÙÙŠ session
@app.route("/meals/delete/<int:idx>", methods=["POST"])
def meals_delete(idx):
    if not require_login():
        return redirect(url_for("login"))
    ensure_today()

    meals = session.get("meals", [])
    if 0 <= idx < len(meals):
        removed = meals.pop(idx)
        session["meals"] = meals

        removed_total = float(removed.get("total", 0.0) or 0.0)
        session["consumed"] = max(0.0, float(session.get("consumed", 0.0)) - removed_total)

    return redirect(url_for("diet_dashboard"))

# ---------- WORKOUT DASHBOARD ----------
@app.route("/workout")
def workout_dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    day = session.get("day")

    burned = get_workouts_total(session["user_id"], day)
    workouts = get_workouts(session["user_id"], day)

    return render_template(
        "workout_dashboard.html",
        user_name=session.get("user_name"),
        burned=round(burned, 2),
        workouts=workouts,
        day=day
    )

@app.route("/workouts/add", methods=["POST"])
def workouts_add():
    if not require_login():
        return redirect(url_for("login"))
    ensure_today()

    wtype = request.form.get("workout_type","walking")
    minutes = float(request.form.get("minutes","0") or 0)
    day = session.get("day")

    weight = get_user_weight(session["user_id"])
    calories = float(round(estimate_calories(wtype, minutes, weight), 2))

    add_workout(session["user_id"], day, wtype, minutes, calories)
    return redirect(url_for("workout_dashboard"))

# âœ… Ø­Ø°Ù ØªÙ…Ø±ÙŠÙ† Ù…Ù† DB Ø¨Ø§Ù„Ù€ id
@app.route("/workouts/delete/<int:wid>", methods=["POST"])
def workouts_delete(wid):
    if not require_login():
        return redirect(url_for("login"))
    delete_workout(session["user_id"], wid)
    return redirect(url_for("workout_dashboard"))

# ---------- MEALS FLOW ----------
@app.route("/start/<meal>")
def start_meal(meal):
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    meal = (meal or "").lower().strip()
    allowed = {"breakfast","lunch","dinner","snack"}
    if meal not in allowed:
        meal = "breakfast"

    session["current_meal"] = meal
    return redirect(url_for("upload"))

@app.route("/upload", methods=["GET","POST"])
def upload():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    meal = session.get("current_meal","breakfast")

    if request.method == "POST":
        # âœ… accept camera OR gallery
        img = None
        if "image_gallery" in request.files and request.files["image_gallery"].filename != "":
            img = request.files["image_gallery"]
        elif "image" in request.files and request.files["image"].filename != "":
            img = request.files["image"]

        if img is None:
            return render_template("upload.html", error="Ø§Ø±ÙØ¹ ØµÙˆØ±Ø© Ø£Ùˆ ØµÙˆÙ‘Ø± ØµÙˆØ±Ø© Ù…Ù† Ø§Ù„ÙƒØ§Ù…ÙŠØ±Ø§.", meal_type=meal)

        image_path = "/content/static/uploaded.jpg"
        img.save(image_path)

        results = model(image_path)
        df = results.pandas().xyxy[0]

        foods_raw = df["name"].unique().tolist() if len(df) else []
        foods = [norm_label(x) for x in foods_raw]

        annotated_url = None
        try:
            results.save()
            exp_dirs = sorted(glob.glob("runs/detect/exp*"), key=os.path.getmtime)
            latest = exp_dirs[-1] if exp_dirs else None
            if latest:
                imgs = glob.glob(os.path.join(latest, "*.jpg")) + glob.glob(os.path.join(latest, "*.png"))
                if imgs:
                    shutil.copy(imgs[0], "/content/static/annotated.jpg")
                    annotated_url = "/static/annotated.jpg"
        except:
            annotated_url = None

        res = []
        for k in foods:
            if k in calorie_data:
                res.append(f"{label_display.get(k, k)}: {calorie_data[k]} kcal / 100g")
            else:
                res.append(f"{label_display.get(k, k)}: Ù…Ø´ Ù…ÙˆØ¬ÙˆØ¯ ÙÙŠ Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª")

        return render_template(
            "results.html",
            foods=foods,
            res=res,
            annotated_url=annotated_url,
            daily_goal=session["daily_goal"],
            meal_type=meal
        )

    return render_template("upload.html", meal_type=meal)

@app.route("/calculate", methods=["POST"])
def calculate():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    foods = request.form.getlist("food")
    qtys  = request.form.getlist("quantity")

    total = 0.0
    details = []

    for f, q in zip(foods, qtys):
        key = norm_label(f)
        q = float(q)

        cal_per_100 = calorie_data.get(key, 0.0)
        cal = cal_per_100 * (q / 100.0)

        total += cal
        details.append({
            "food": label_display.get(key, key),
            "qty": q,
            "cal": round(cal, 2) if cal_per_100 > 0 else None
        })

    total = round(total, 2)

    session["meals"] = session.get("meals", [])
    session["meals"].append({
        "meal_type": session.get("current_meal","meal"),
        "total": total,
        "details": details
    })

    session["consumed"] = float(session.get("consumed", 0.0)) + total
    return redirect(url_for("diet_dashboard"))

# ================== RUN (colab) ==================
def run_app():
    os.makedirs("/content/static", exist_ok=True)
    ngrok.kill()
    ngrok.set_auth_token(NGROK_TOKEN)
    public_url = ngrok.connect(5000)
    print("âœ… Ø§ÙØªØ­ Ø§Ù„Ù„ÙŠÙ†Ùƒ Ø¯Ù‡:", public_url)
    app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)

if __name__ == "__main__":
    run_app()
```

```python
%%writefile /content/app.py
from flask import Flask, request, render_template, redirect, url_for, session
import torch
import csv
import os
import glob
import shutil
from datetime import date
import sqlite3
from werkzeug.security import generate_password_hash, check_password_hash
from pyngrok import ngrok

# ================== CONFIG ==================
MODEL_PATH = "/content/last (2).pt"
CSV_PATH   = "/content/caloris.csv"
DB_PATH    = "/content/app.db"
NGROK_TOKEN = "35c97fFqjssPXEyCdZm46xBofrg_2Fxk6xyfyFYnLgaGUe8qb"  # ğŸ‘ˆ Ø­Ø· ØªÙˆÙƒÙ† Ø¬Ø¯ÙŠØ¯

# ================== APP ==================
app = Flask(__name__)
app.secret_key = "change-this-secret"

# ================== NORMALIZE LABELS ==================
def norm_label(s: str) -> str:
    s = (s or "").strip().lower()
    s = s.replace("_", " ")
    s = " ".join(s.split())
    return s

# ================== DATABASE ==================
def db_conn():
    return sqlite3.connect(DB_PATH)

def init_db():
    con = db_conn()
    cur = con.cursor()

    cur.execute(
        "CREATE TABLE IF NOT EXISTS users ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "name TEXT NOT NULL,"
        "email TEXT NOT NULL UNIQUE,"
        "password_hash TEXT NOT NULL"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS profiles ("
        "user_id INTEGER PRIMARY KEY,"
        "sex TEXT,"
        "age INTEGER,"
        "height_cm REAL,"
        "weight_kg REAL,"
        "activity TEXT,"
        "goal TEXT,"
        "daily_goal REAL,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    cur.execute(
        "CREATE TABLE IF NOT EXISTS workouts ("
        "id INTEGER PRIMARY KEY AUTOINCREMENT,"
        "user_id INTEGER NOT NULL,"
        "day TEXT NOT NULL,"
        "workout_type TEXT NOT NULL,"
        "minutes REAL NOT NULL,"
        "calories REAL NOT NULL,"
        "created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,"
        "FOREIGN KEY(user_id) REFERENCES users(id)"
        ")"
    )

    con.commit()
    con.close()

init_db()

def get_user_by_email(email: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT id, name, email, password_hash FROM users WHERE email=?", (email,))
    row = cur.fetchone()
    con.close()
    return row

def create_user(name: str, email: str, password: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO users (name,email,password_hash) VALUES (?,?,?)",
        (name, email, generate_password_hash(password))
    )
    con.commit()
    uid = cur.lastrowid
    con.close()
    return uid

def upsert_profile(uid: int, sex: str, age: int, height: float, weight: float, activity: str, goal: str, daily: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO profiles VALUES (?,?,?,?,?,?,?,?) "
        "ON CONFLICT(user_id) DO UPDATE SET "
        "sex=excluded.sex, age=excluded.age, height_cm=excluded.height_cm, "
        "weight_kg=excluded.weight_kg, activity=excluded.activity, "
        "goal=excluded.goal, daily_goal=excluded.daily_goal",
        (uid, sex, age, height, weight, activity, goal, daily)
    )
    con.commit()
    con.close()

def get_daily_goal(uid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT daily_goal FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return row[0] if row else None

def get_user_weight(uid: int) -> float:
    con = db_conn()
    cur = con.cursor()
    cur.execute("SELECT weight_kg FROM profiles WHERE user_id=?", (uid,))
    row = cur.fetchone()
    con.close()
    return float(row[0]) if row and row[0] else 70.0

# ----- workouts db -----
def add_workout(uid: int, day: str, workout_type: str, minutes: float, calories: float):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "INSERT INTO workouts (user_id, day, workout_type, minutes, calories) VALUES (?,?,?,?,?)",
        (uid, day, workout_type, minutes, calories)
    )
    con.commit()
    con.close()

def get_workouts(uid: int, day: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT workout_type, minutes, calories, id FROM workouts WHERE user_id=? AND day=? ORDER BY id DESC",
        (uid, day)
    )
    rows = cur.fetchall()
    con.close()
    return rows

def get_workouts_total(uid: int, day: str):
    con = db_conn()
    cur = con.cursor()
    cur.execute(
        "SELECT COALESCE(SUM(calories),0) FROM workouts WHERE user_id=? AND day=?",
        (uid, day)
    )
    total = cur.fetchone()[0]
    con.close()
    return float(total or 0.0)

def delete_workout(uid: int, wid: int):
    con = db_conn()
    cur = con.cursor()
    cur.execute("DELETE FROM workouts WHERE id=? AND user_id=?", (wid, uid))
    con.commit()
    con.close()

# ================== MODEL ==================
model = torch.hub.load(
    "ultralytics/yolov5",
    "custom",
    path=MODEL_PATH,
    trust_repo=True
)

# ================== LOAD CALORIES CSV ==================
calorie_data = {}
label_display = {}

with open(CSV_PATH, "r", encoding="cp1256", newline="") as f:
    reader = csv.reader(f)
    for row in reader:
        if not row or len(row) < 3:
            continue
        raw_name = row[0].strip()
        key = norm_label(raw_name)
        try:
            calorie_data[key] = float(row[2])
            label_display[key] = raw_name
        except:
            pass

# ================== CALC DAILY GOAL ==================
def calc_daily(sex, age, height, weight, activity, goal):
    sex = (sex or "").lower().strip()
    if sex == "male":
        bmr = 10*weight + 6.25*height - 5*age + 5
    else:
        bmr = 10*weight + 6.25*height - 5*age - 161

    factors = {"sedentary":1.2,"light":1.375,"moderate":1.55,"active":1.725,"very_active":1.9}
    tdee = bmr * factors.get(activity, 1.2)

    if goal == "lose":
        tdee -= 500
    elif goal == "gain":
        tdee += 300

    tdee = max(tdee, 1200 if sex == "female" else 1500)
    return float(round(tdee, 0))

# ================== SESSION ==================
def ensure_today():
    today = str(date.today())
    if session.get("day") != today:
        session["day"] = today
        session["meals"] = []
        session["consumed"] = 0.0

def require_login():
    return "user_id" in session

# ================== WORKOUT CAL ESTIMATE ==================
def estimate_calories(workout_type: str, minutes: float, weight_kg: float):
    mets = {
        # Cardio
        "walking": 3.5, "running": 8.0, "cycling": 6.0, "swimming": 7.0,
        "jump_rope": 10.0, "elliptical": 5.0, "stair_climber": 8.8,
        "rowing": 7.0, "hiit": 9.0, "football": 7.0, "basketball": 8.0,
        # Gym
        "weights_light": 3.0, "weights_moderate": 5.0, "weights_heavy": 6.0,
        "circuit_training": 8.0, "crossfit": 10.0,
    }
    met = mets.get(workout_type, 4.0)
    return met * 3.5 * weight_kg / 200.0 * minutes

# ================== ROUTES ==================
@app.route("/")
def root():
    if not require_login():
        return redirect(url_for("login"))
    daily = get_daily_goal(session["user_id"])
    if daily is None:
        return redirect(url_for("onboarding"))
    session["daily_goal"] = float(daily)
    ensure_today()
    return redirect(url_for("home"))

@app.route("/home")
def home():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    uid = session["user_id"]
    day = session.get("day")

    daily_goal = float(session["daily_goal"])
    consumed = float(session.get("consumed", 0.0))
    burned = float(get_workouts_total(uid, day))

    # ØµØ§ÙÙŠ Ø§Ù„ÙŠÙˆÙ… = Ø£ÙƒÙ„ - ØªÙ…Ø§Ø±ÙŠÙ†
    net = max(0.0, consumed - burned)

    remaining_net = round(daily_goal - net, 2)
    if remaining_net >= 0:
        # ÙƒÙˆÙŠØ³ - ÙˆÙ„Ø³Ù‡ Ø¨Ø§Ù‚ÙŠ
        need_burn_or_remaining = round(remaining_net, 2)
        # ØªØ­Ø°ÙŠØ± Ø¨Ø³ÙŠØ· Ù„Ùˆ Ù‚Ø±Ø¨ ÙŠØ®Ù„Øµ (Ø£Ù‚Ù„ Ù…Ù† 10% Ù…Ù† Ø§Ù„Ù‡Ø¯Ù)
        status_type = "warn" if (need_burn_or_remaining <= (daily_goal * 0.10)) else "good"
    else:
        # Ø¹Ø¯ÙŠØª - Ù…Ø­ØªØ§Ø¬ ØªØ­Ø±Ù‚
        need_burn_or_remaining = round(abs(remaining_net), 2)
        remaining_net = 0.0
        status_type = "bad"

    return render_template(
        "home.html",
        day=day,
        daily_goal=round(daily_goal, 0),
        consumed=round(consumed, 2),
        burned=round(burned, 2),
        net=round(net, 2),
        remaining_net=round(remaining_net, 2),
        status_type=status_type,
        need_burn_or_remaining=need_burn_or_remaining
    )

@app.route("/go/diet")
def go_diet():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("diet_dashboard"))

@app.route("/go/workout")
def go_workout():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("workout_dashboard"))

@app.route("/recalc")
def recalc():
    if not require_login():
        return redirect(url_for("login"))
    return redirect(url_for("onboarding"))

# ---------- AUTH ----------
@app.route("/signup", methods=["GET","POST"])
def signup():
    if request.method == "POST":
        name = request.form.get("name","").strip()
        email = request.form.get("email","").strip().lower()
        password = request.form.get("password","")

        if not name or not email or not password:
            return render_template("signup.html", error="Ø§Ù…Ù„Ø£ ÙƒÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª.")
        if get_user_by_email(email):
            return render_template("signup.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ Ù…ÙˆØ¬ÙˆØ¯ Ø¨Ø§Ù„ÙØ¹Ù„. Ø³Ø¬Ù„ Ø¯Ø®ÙˆÙ„.")

        uid = create_user(name, email, password)
        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = email
        return redirect(url_for("onboarding"))

    return render_template("signup.html")

@app.route("/login", methods=["GET","POST"])
def login():
    if request.method == "POST":
        email = request.form.get("email","").strip().lower()
        password = request.form.get("password","")

        row = get_user_by_email(email)
        if not row:
            return render_template("login.html", error="Ø§Ù„Ø¥ÙŠÙ…ÙŠÙ„ ØºÙŠØ± ØµØ­ÙŠØ­.")
        uid, name, em, pw_hash = row
        if not check_password_hash(pw_hash, password):
            return render_template("login.html", error="ÙƒÙ„Ù…Ø© Ø§Ù„Ù…Ø±ÙˆØ± ØºÙŠØ± ØµØ­ÙŠØ­Ø©.")

        session.clear()
        session["user_id"] = uid
        session["user_name"] = name
        session["user_email"] = em
        return redirect(url_for("root"))

    return render_template("login.html")

@app.route("/logout")
def logout():
    session.clear()
    return redirect(url_for("login"))

# ---------- ONBOARDING ----------
@app.route("/onboarding", methods=["GET","POST"])
def onboarding():
    if not require_login():
        return redirect(url_for("login"))

    if request.method == "POST":
        sex = request.form.get("sex")
        age = int(request.form.get("age"))
        height = float(request.form.get("height_cm"))
        weight = float(request.form.get("weight_kg"))
        activity = request.form.get("activity")
        goal = request.form.get("goal")

        daily = calc_daily(sex, age, height, weight, activity, goal)
        upsert_profile(session["user_id"], sex, age, height, weight, activity, goal, daily)

        session["daily_goal"] = daily
        ensure_today()
        return redirect(url_for("home"))

    return render_template("onboarding.html")

# ---------- DIET DASHBOARD ----------
@app.route("/diet")
def diet_dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    daily = float(session["daily_goal"])
    consumed = float(session.get("consumed", 0.0))

    remaining = max(0.0, round(daily - consumed, 2))
    over_by = max(0.0, round(consumed - daily, 2))

    return render_template(
        "diet_dashboard.html",
        user_name=session.get("user_name"),
        daily_goal=round(daily, 0),
        consumed=round(consumed, 2),
        remaining=remaining,
        over_by=over_by,
        meals=session.get("meals", []),
        day=session.get("day")
    )

@app.route("/meals/delete/<int:idx>", methods=["POST"])
def meals_delete(idx):
    if not require_login():
        return redirect(url_for("login"))
    ensure_today()

    meals = session.get("meals", [])
    if 0 <= idx < len(meals):
        removed = meals.pop(idx)
        session["meals"] = meals
        removed_total = float(removed.get("total", 0.0) or 0.0)
        session["consumed"] = max(0.0, float(session.get("consumed", 0.0)) - removed_total)

    return redirect(url_for("diet_dashboard"))

# ---------- WORKOUT DASHBOARD ----------
@app.route("/workout")
def workout_dashboard():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    uid = session["user_id"]
    day = session.get("day")

    burned = get_workouts_total(uid, day)
    workouts = get_workouts(uid, day)

    return render_template(
        "workout_dashboard.html",
        user_name=session.get("user_name"),
        burned=round(burned, 2),
        workouts=workouts,
        day=day
    )

@app.route("/workouts/add", methods=["POST"])
def workouts_add():
    if not require_login():
        return redirect(url_for("login"))
    ensure_today()

    wtype = request.form.get("workout_type","walking")
    minutes = float(request.form.get("minutes","0") or 0)
    day = session.get("day")

    weight = get_user_weight(session["user_id"])
    calories = float(round(estimate_calories(wtype, minutes, weight), 2))

    add_workout(session["user_id"], day, wtype, minutes, calories)
    return redirect(url_for("workout_dashboard"))

@app.route("/workouts/delete/<int:wid>", methods=["POST"])
def workouts_delete(wid):
    if not require_login():
        return redirect(url_for("login"))
    delete_workout(session["user_id"], wid)
    return redirect(url_for("workout_dashboard"))

# ---------- MEALS FLOW ----------
@app.route("/start/<meal>")
def start_meal(meal):
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    meal = (meal or "").lower().strip()
    allowed = {"breakfast","lunch","dinner","snack"}
    if meal not in allowed:
        meal = "breakfast"

    session["current_meal"] = meal
    return redirect(url_for("upload"))

@app.route("/upload", methods=["GET","POST"])
def upload():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()
    meal = session.get("current_meal","breakfast")

    if request.method == "POST":
        img = None
        if "image_gallery" in request.files and request.files["image_gallery"].filename != "":
            img = request.files["image_gallery"]
        elif "image" in request.files and request.files["image"].filename != "":
            img = request.files["image"]

        if img is None:
            return render_template("upload.html", error="Ø§Ø±ÙØ¹ ØµÙˆØ±Ø© Ø£Ùˆ ØµÙˆÙ‘Ø± ØµÙˆØ±Ø© Ù…Ù† Ø§Ù„ÙƒØ§Ù…ÙŠØ±Ø§.", meal_type=meal)

        os.makedirs("/content/static", exist_ok=True)
        image_path = "/content/static/uploaded.jpg"
        img.save(image_path)

        results = model(image_path)
        df = results.pandas().xyxy[0]

        foods_raw = df["name"].unique().tolist() if len(df) else []
        foods = [norm_label(x) for x in foods_raw]

        annotated_url = None
        try:
            results.save()
            exp_dirs = sorted(glob.glob("runs/detect/exp*"), key=os.path.getmtime)
            latest = exp_dirs[-1] if exp_dirs else None
            if latest:
                imgs = glob.glob(os.path.join(latest, "*.jpg")) + glob.glob(os.path.join(latest, "*.png"))
                if imgs:
                    shutil.copy(imgs[0], "/content/static/annotated.jpg")
                    annotated_url = "/static/annotated.jpg"
        except:
            annotated_url = None

        res = []
        for k in foods:
            if k in calorie_data:
                res.append(f"{label_display.get(k, k)}: {calorie_data[k]} kcal / 100g")
            else:
                res.append(f"{label_display.get(k, k)}: Ù…Ø´ Ù…ÙˆØ¬ÙˆØ¯ ÙÙŠ Ù…Ù„Ù Ø§Ù„Ø³Ø¹Ø±Ø§Øª")

        return render_template(
            "results.html",
            foods=foods,
            res=res,
            annotated_url=annotated_url,
            daily_goal=session["daily_goal"],
            meal_type=meal
        )

    return render_template("upload.html", meal_type=meal)

@app.route("/calculate", methods=["POST"])
def calculate():
    if not require_login():
        return redirect(url_for("login"))
    if "daily_goal" not in session:
        return redirect(url_for("onboarding"))

    ensure_today()

    foods = request.form.getlist("food")
    qtys  = request.form.getlist("quantity")

    total = 0.0
    details = []

    for f, q in zip(foods, qtys):
        key = norm_label(f)
        q = float(q)
        cal_per_100 = calorie_data.get(key, 0.0)
        cal = cal_per_100 * (q / 100.0)

        total += cal
        details.append({
            "food": label_display.get(key, key),
            "qty": q,
            "cal": round(cal, 2) if cal_per_100 > 0 else None
        })

    total = round(total, 2)

    session["meals"] = session.get("meals", [])
    session["meals"].append({
        "meal_type": session.get("current_meal","meal"),
        "total": total,
        "details": details
    })

    session["consumed"] = float(session.get("consumed", 0.0)) + total
    return redirect(url_for("diet_dashboard"))

# ================== RUN (colab) ==================
def run_app():
    os.makedirs("/content/static", exist_ok=True)
    ngrok.kill()
    ngrok.set_auth_token(NGROK_TOKEN)
    public_url = ngrok.connect(5000)
    print("âœ… Ø§ÙØªØ­ Ø§Ù„Ù„ÙŠÙ†Ùƒ Ø¯Ù‡:", public_url)
    app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)

if __name__ == "__main__":
    run_app()
```
